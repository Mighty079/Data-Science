```python
# Loop PS q1:multiplication table
print("Enter number:")
a = int(input())
for i in range(1,11):
    print(a,"X",i,"=", a*i)
```

```python
# Loop PS q2: FInding prime number
a=int(input("Enter a number:"))

divisors=[]
for i in range(2,a):
    if a%i==0:
        divisors.append(i)
if len(divisors)==0:
    print("Prime")
else:
    print("Not Prime")
    print("Divisible by",divisors)
```

```python
#Sum of n natural numbers
n = int(input("Enter the number:"))
sum = 0
for i in range(1,n+1):
        sum = sum + i 
print(sum)
    
```

```python
#Factorial of a number
n = int(input("Enter the number:"))
sum = 1
for i in range(1,n+1):
        sum = sum * i 
print("Factorial of",n,":",sum)
    
```
